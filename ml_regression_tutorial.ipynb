## Data Generation

```python
%matplotlib inline
```

```python
import numpy as np
from sklearn.svm import SVR
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt


# #############################################################################
# Generate sample data
N = 120
np.random.seed(42)
Xpre = np.sort(6 * np.random.rand(N, 1), axis=0)
X1 = Xpre-3
X = X1.reshape(120,1)
y = np.cos(X).ravel()
Xtest = 3 * np.arange(-1,1,0.01)
Xtest1 = Xtest.reshape(200,1)

# #############################################################################
# Add noise to targets
# 'Outliers'
y[::5] += 3 * (0.5 - np.random.rand(24))
```

## Train regressors

### Support Vector Regression

```python
# Support Vector regression
svr_rbf = SVR(kernel='rbf', C=10, gamma=0.1)
svr_rbf.fit(X, y)
y_rbf = svr_rbf.predict(Xtest1)
```

### Decision Tree Regression 

# Decision Tree Regression
regr_dtr = tree.DecisionTreeRegressor(max_depth=3)
regr_dtr.fit(X, y)
y_dtr = regr_dtr.predict(Xtest1)

### Random Forest regression

# Random Forest Regression
regr_rf = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=2)
regr_rf.fit(X, y)
y_rf = regr_rf.predict(Xtest1)

### Artificial Neural Network (Regression)

```python
# Keras (with Tensorflow)
# load modules
from keras import optimizers
from keras.models import Sequential
from keras.layers import Dense, Activation

# DIREKT PROBLEM -----------------------------------------------------------

# model definition
model = Sequential()
model.add(Dense(10, input_dim=1, activation='sigmoid'))
model.add(Dense(50, activation='sigmoid'))
model.add(Dense(1, activation='linear'))

# compile the model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse', 'mae'])

# fit the model to the dataset
model.fit(X, y, epochs=1000, batch_size=10)
y_nn = model.predict(Xtest1)
```

```python
# INVERSE PROBLEM -----------------------------------------------------------
# model definition
model_2 = Sequential()
model_2.add(Dense(10, input_dim=1, activation='sigmoid'))
model_2.add(Dense(50, activation='sigmoid'))
model_2.add(Dense(1, activation='linear'))

# compile the model
model_2.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse', 'mae'])

# fit the model to the dataset
model_2.fit(y, X, epochs=1000, batch_size=10)
y_nn_inverse = model_2.predict(Xtest1)
```

## Analyze and Visualize Results

```python
# Reshape data
y_nn5 = y_nn.reshape(1, 200)
y_nn5_inverse = y_nn_inverse.reshape(1, 200)

# Calculate errors
y_expect = np.cos(Xtest)
chi2_svr = np.std(y_rbf - y_expect)
#chi2_dtr = np.std(y_dtr - y_expect)
#chi2_rf = np.std(y_rf - y_expect)
chi2_nn = np.std(y_nn5 - y_expect)
chi2_nn_inverse = np.std(y_nn5_inverse - y_expect)
print('Standard deviations')
print('SVR: ',chi2_svr)
#print('DTR: ',chi2_dtr)
#print('RFR: ',chi2_rf)
print('ANN: ',chi2_nn)
print('Inverse-ANN: ',chi2_nn_inverse)

# #############################################################################
# Look at the results
lw = 3
plt.figure(figsize=(10, 8))
plt.axis([-3, 3, -3, 3]) # axes limits ([xmin, xmax, ymin, ymax])
plt.scatter(X, y, color='darkorange', label='Data')
plt.scatter(y, X, color='blue', label='Inverse data')
plt.plot(Xtest, y_expect, color='red', label='Expectation')
plt.plot(Xtest, y_rbf, color='navy', lw=lw, label='Support vector')
#plt.plot(Xtest, y_dtr, color='c', lw=lw, label='Decision tree')
#plt.plot(Xtest, y_rf, color='cornflowerblue', lw=lw, label='Random forest')
plt.plot(Xtest, y_nn, color='black', lw=lw, label='Neural network')
plt.plot(Xtest, y_nn_inverse, color='c', lw=lw, label='Inverse neural network')

plt.grid(which='both', axis='both')
plt.xlabel('Data')
plt.ylabel('Target')
plt.title('Machine Learning Regressions')
plt.legend()
plt.show()
```

## Fine-tunig the model

### SVR

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(svr_rbf, X, y, scoring="neg_mean_squared_error", cv=10)
svr_rbf_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(svr_rbf_scores)
```

### Grid Search

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
        {'kernel': ['linear'], 'C': [1.0, 10., 30., 100., 300., 1000., 3000., 10000., 30000.0]},
        {'kernel': ['rbf'], 'C': [1.0, 3.0, 4.0, 5.0, 6.0, 7.0, 10., 20.0, 100., 500., 1000.0, 5000.0, 10000.0],
         'gamma': [0.01, 0.03, 0.1, 0.3, 1.0, 3.0]},
    ]

svm_reg = SVR()
grid_search = GridSearchCV(svm_reg, param_grid, cv=5, scoring='neg_mean_squared_error', verbose=2, n_jobs=4)
grid_search.fit(X, y)
```

```python
negative_mse = grid_search.best_score_
rmse = np.sqrt(-negative_mse)
rmse
```

```python
grid_search.best_params_
```

### Randomized Search

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import expon, reciprocal

# see https://docs.scipy.org/doc/scipy/reference/stats.html
# for `expon()` and `reciprocal()` documentation and more probability distribution functions.

# Note: gamma is ignored when kernel is "linear"
param_distribs = {
        'kernel': ['linear', 'rbf'],
        'C': reciprocal(1, 200000),
        'gamma': expon(scale=1.0),
    }

svm_reg = SVR()
rnd_search = RandomizedSearchCV(svm_reg, param_distributions=param_distribs,
                                n_iter=50, cv=5, scoring='neg_mean_squared_error',
                                verbose=2, n_jobs=4, random_state=42)
rnd_search.fit(X, y)
```

```python
negative_mse = rnd_search.best_score_
rmse = np.sqrt(-negative_mse)
rmse
```

```python
rnd_search.best_params_
```
